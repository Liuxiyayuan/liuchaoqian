```python
import matplotlib.pyplot as plt
import numpy as np
#以下是对图像横纵坐标的制定
x = np.linspace(0,2 * np.pi,200)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show
```

```python

#坐标制定,此时没处理不可使用中文（因为默认不支持中文字符）
x = [1,2,3,4]
y = [1,5,6,8]
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```

```python
plt.rcParams['font.sans-serif']= ['SimHei']#处理字体，让其能支持中文字符
plt.rcParams['axes.unicode_minus']=False#正常显示负号
year = [2005,2006,2007,2008,2009,2010]
people = [18,31,40,39,36,39]
#生成图表
plt.plot(year,people)
plt.xlabel('年份')#横坐标
plt.ylabel('人口')#纵坐标
plt.title('人口增长')
#设置纵坐标刻度
plt.yticks([0,18,31,40,39,36,39])
#设置填充选项：参数分别对应横坐标，纵坐标，纵坐标填充起始值，填充颜色
plt.fill_between(year,people,18,color = 'purple')
#调用函数图表使之出现
plt.show()
```
