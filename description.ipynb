```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
df = pd.read_spss(R'data\identity.sav')
```

```python
df['民族'].value_counts(normalize=True)
```

```python
x = df['民族'].value_counts().index
y = df['民族'].value_counts(normalize=True)
fig, ax2 = plt.subplots()
labels = df['民族'].value_counts().index
ax2.pie(y,autopct='%1.1f%%',labels= labels,textprops=dict(color='w'))
plt.show()
```

```python
#有序定类变量描述统计，可增加累计次数和累计比例的统计。可采用饼图或柱状图
import pyreadstat 
df['年级'].value_counts()
```

```python
df2,metadata = pyreadstat.read_sav(R'data\identity.sav',apply_value_formats=True,formats_as_ordered_category=True)
result = df['会打多少分'].value_counts(sort=False)
```

```python
df2['年级'].value_counts(sort=False)
```

```python
result = df2['年级'].value_counts(sort=False)
df_result = pd.DataFrame(result)
df_result['累计求和'] = result.values.cumsum()
#df_result['新变量'] = 1 为表格增加新变量
#df_result['再增新变量'] = df_result.新变量+1
df_result['比例'] = df2['年级'].value_counts(normalize=True)
```

```python
x = df_result.index
y = df_result['年级'].values
fig, ax2 = plt.subplots()
ax2.bar(x,y)
plt.show()
```
